```python
import networkx as nx
import matplotlib.pyplot as plt
from nanonets import NanoparticleTopology
```

# Lattice Networks

```python
# Parameter and Inti
N_x, N_y        = 5,3
electrode_pos   = [[0,0],[2,0],[0,2],[2,2]]
lattice_net     = NanoparticleTopology()

# Build Network and attach Electrodes
lattice_net.lattice_network(N_x, N_y)
lattice_net.add_electrodes_to_lattice_net(electrode_pos)

# Extract Graph & Positions
G   = lattice_net.get_graph()
pos = lattice_net.get_positions()

print(lattice_net)
fig, ax = plt.subplots(dpi=150)
nx.draw_networkx(G=G, pos=pos, ax=ax)
```

# Random Networks

```python
# Parameter and Inti
N_particles     = 20
electrode_pos   = [[-1,-1],[-1,1],[1,-1],[1,1]]
rng_net         = NanoparticleTopology()

# Build Network and attach Electrodes
rng_net.random_network(N_particles)
rng_net.add_electrodes_to_random_net(electrode_pos)

# Extract Graph & Positions
G   = rng_net.get_graph()
pos = rng_net.get_positions()

print(rng_net)
fig, ax = plt.subplots(dpi=150)
nx.draw_networkx(G=G, pos=pos, ax=ax)
```

# Add a NP to the Output Electrode

```python
# Parameter and Inti
N_x, N_y        = 5,3
electrode_pos   = [[0,0],[2,0],[0,2],[2,2]]
lattice_net     = NanoparticleTopology()

# Build Network and attach Electrodes
lattice_net.lattice_network(N_x, N_y)
lattice_net.add_electrodes_to_lattice_net(electrode_pos)
lattice_net.add_np_to_output()

# Extract Graph & Positions
G   = lattice_net.get_graph()
pos = lattice_net.get_positions()

print(lattice_net)
fig, ax = plt.subplots(dpi=150)
nx.draw_networkx(G=G, pos=pos, ax=ax)
```
