```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
```

```python
def self_capacitance(radius, eps):

    return 4*np.pi*eps*radius

def mutual_capacitance_adjacent_spheres(radius1, radius2, spacing, eps, order=100):

    distance    = radius1 + radius2 + spacing
    factor      = 2*np.pi*eps*radius1*radius2/distance

    sum_value   = [(radius1**n + radius2**n)/(distance**n) for n in range(order)]

    return factor*np.sum(sum_value)
```

```python
radius1 = 10
radius2 = 10
spacing = 1
eps_mol = 2.6
eps_ins = 3.9

orders  = np.arange(0,51)
C_ij    = [mutual_capacitance_adjacent_spheres(radius1, radius2, spacing, eps=eps_mol, order=order) for order in orders]
```

```python
with plt.style.context(["science"]):
    
    fig = plt.figure(dpi=200)
    ax  = fig.add_subplot()
    ax.plot(orders, C_ij, '.-')
    ax.set_ylabel("$C_{ij}^{(N)}$")
    ax.set_xlabel("$N$")
```
