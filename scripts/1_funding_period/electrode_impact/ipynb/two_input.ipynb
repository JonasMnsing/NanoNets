```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scienceplots

plt.style.use(["science","bright","grid"])
```

```python
path    = "/mnt/c/Users/jonas/Desktop/phd/data/1_funding_period/electrode_impact/voltage_sweep/"
N_vals  = np.arange(3,10)
data    = {n : pd.read_csv(f"{path}Nx={n}_Ny={n}_Ne=8_two_input.csv").loc[:,['E1','E3','Observable','Error']] for n in N_vals}
data[3]
```

```python
n   = 9
x   = data[n]['E1'].values*1000
y   = data[n]['Observable'].values*1e-9
y_e = data[n]['Error'].values*1e-9

fig = plt.figure(dpi=200)
ax  = fig.add_subplot()
ax.errorbar(x, y)
ax.fill_between(x, y+y_e, y-y_e, color='grey')
ax.set_xlabel("$U$ [mV]")
ax.set_ylabel("$I$ [nA]")
```
