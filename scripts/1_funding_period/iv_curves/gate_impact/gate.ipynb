```python
import matplotlib.pyplot as plt
import scienceplots
import pandas as pd
```

```python
df = pd.read_csv("Nx=5_Ny=5_Nz=1_Ne=8.csv")
df = df[df['Error'] != 0]

df['G'] = df['G']*1000
```

```python
cmeter      = 1/2.54
single_col  = 8.6

with plt.style.context(["science","bright","grid"]):

    fig = plt.figure()
    fig.set_figwidth(2.5*single_col*cmeter/2)
    fig.set_figheight(single_col*cmeter/1.5)
    ax  = fig.add_subplot()
    ax.fill_between(x=df['G'], y1=df['Current']-df['Error'], y2=df['Current']+df['Error']) 
    ax.set_xlabel("$U_G$ [mV]")
    ax.set_ylabel("$I$ [pA]")
    ax.set_xlim(-100,100)
    ax.set_ylim(-750,-350)

fig.savefig("gate_impact.svg", bbox_inches='tight', transparent=True)
fig.savefig("gate_impact.pdf", bbox_inches='tight', transparent=False)
plt.show()
```
