### Surrogate Model Data

```python
# --- Configuration ---
SCALE_FACTOR = 0.16 

# 1. Load artifacts for Epoch 50
model = tf.keras.models.load_model('../../../../trained_model_epoch_200.keras')

with open('../../../../scalers.pkl', 'rb') as f:
    scalers = pickle.load(f)
    x_scaler = scalers['x_scaler']
    y_scaler = scalers['y_scaler']

with open('../../../../trained_model_history_epoch_200.pkl', 'rb') as f:
    history = pickle.load(f)

# 2. Load test data
data_test = np.load('../../../../test_data.npz')
X_test_scaled = data_test['X_test_scaled']
y_test_trans = data_test['y_test'] 

# 3. Predict in scaled neural network space [-1, 1]
y_pred_scaled = model.predict(X_test_scaled)

# 4. Step A: Inverse MinMax Scaling (Back to arcsinh space)
y_pred_trans = y_scaler.inverse_transform(y_pred_scaled)

# 5. Step B: Physics Recovery (Un-arcsinh back to physical aA)
y_pred_phys = SCALE_FACTOR * np.sinh(y_pred_trans)
y_test_phys = SCALE_FACTOR * np.sinh(y_test_trans)

# --- Quick Verification ---
print("-" * 30)
print(f"Data Points Evaluated: {len(y_test_phys)}")
print(f"Max Predicted Current: {y_pred_phys.max():.2e} aA")
print(f"Max Actual Current:    {y_test_phys.max():.2e} aA")
print(f"Min Actual Current:    {y_test_phys.min():.2e} aA")
print("-" * 30)
```

```python
N_PARTICLES = 9
N_E         = 8
V_ON        = 0.01
V_OFF       = 0.0
V_RANGE     = 0.05
INPUT_POS   = [1,3]
SCALE       = True
INPUT_COL   = [f"E{INPUT_POS[0]}",f"E{INPUT_POS[1]}"]
N_SAMPLES   = 10000
TOPO        = {"Nx": N_PARTICLES, "Ny": N_PARTICLES,
                "e_pos" : [[0,0], [int((N_PARTICLES-1)/2),0], [N_PARTICLES-1,0],
                [0,int((N_PARTICLES-1)/2)], [0,N_PARTICLES-1],
                [N_PARTICLES-1,int((N_PARTICLES)/2)],
                [int((N_PARTICLES)/2),(N_PARTICLES-1)], [N_PARTICLES-1,N_PARTICLES-1]],
                "electrode_type" : ['constant']*N_E}
```

```python
Q_NDR_l = []
Q_NDR_r = []
Q_NLS   = []
for i in range(7):
    for j in range(7):
        if i != j:
            voltages        = logic_gate_sample(V_RANGE, [i,j], N_SAMPLES, TOPO, V_ON, sample_technique='uniform')
            voltages        = voltages[:,:-2]
            voltages_s      = x_scaler.transform(voltages)
            current_data_s  = model.predict(voltages_s)
            current_data    = y_scaler.inverse_transform(current_data_s).flatten()
            data            = pd.DataFrame(np.hstack((voltages,current_data[:,np.newaxis],np.zeros(shape=(len(current_data),1)))),
                                        columns=([f"E{i}" for i in range(7)]+['Observable','Error']))
            arr_nonlinear   = nonlinear_parameter(data,input1_column=f'E{i}',input2_column=f'E{j}')
            Q_NDR_l.append(return_ndr(arr_nonlinear['Ml'].values))
            Q_NDR_r.append(return_ndr(arr_nonlinear['Mr'].values))
            Q_NLS.append(return_nls(arr_nonlinear))
        else:
            Q_NDR_l.append(np.nan)
            Q_NDR_r.append(np.nan)
            Q_NLS.append(np.nan)
```

```python
fig, ax = plt.subplots(dpi=300)
w, h    = fig.get_size_inches()
scale   = 0.75
fig.set_size_inches(w * scale, h * scale)

im  = ax.imshow((np.array(Q_NDR_l).reshape(7,7)+np.array(Q_NDR_r).reshape(7,7))*100/2, cmap='cividis', origin='lower')
_   = ax.set_xticks(np.arange(7))
_   = ax.set_yticks(np.arange(7))
cb  = fig.colorbar(im, ax=ax, label="$Q_\\textrm{NDR}$")
```
