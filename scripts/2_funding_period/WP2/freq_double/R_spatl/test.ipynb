```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Time Scale
step_size   = 1e-10
N_voltages  = 1000
time        = step_size*np.arange(N_voltages)

# Voltages
off_state   = 0.1
on_state    = 0.2
on_t1       = 100
on_t2       = 300
voltages    = np.zeros(shape=(N_voltages,3))

# Input Electrode
voltages[:,0]           = np.repeat(off_state, N_voltages)
voltages[on_t1:on_t2,0] = on_state
```

```python
plt.plot(time, voltages[:,0])
```

```python
# Time Scale
step_size   = 1e-10
amplitude   = 0.2
freq        = 20e7 #200000000
N_voltages  = 2000
time        = step_size*np.arange(N_voltages)

# Voltages
voltages    = np.zeros(shape=(N_voltages,3))

# Input Electrode
voltages[:,0] = amplitude*np.cos(freq*time)
```
