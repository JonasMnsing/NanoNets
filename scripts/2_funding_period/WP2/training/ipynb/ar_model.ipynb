```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots

%matplotlib inline
```

```python
N_voltags   = 5000
phis        = [0.01]
epsilon     = np.random.normal(0,0.1,N_voltags)
x_vals      = np.zeros(N_voltags)

for i in range(len(phis), N_voltags):
    x_vals[i]   =   epsilon[i]
    for j, phi in enumerate(phis):
        x_vals  +=  phi*x_vals[i-j]

y_vals  = x_vals[1:]
x_vals  = x_vals[:-1]
```

```python
with plt.style.context(["science","bright"]):

    fig = plt.figure(dpi=200)
    ax  = fig.add_subplot()
    ax.plot(x_vals)
```
