```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
```

```python
N_voltags   = 10000
phis        = [0.1]
epsilon     = np.random.normal(0,0.1,N_voltags)
x_vals      = np.zeros(N_voltags)

for i in range(1, N_voltags):
    x_vals[i]   = phis[0]*x_vals[i-1] + epsilon[i]

y_vals  = x_vals[1:]
x_vals  = x_vals[:-1]
```

```python
with plt.style.context(["science","bright"]):

    fig = plt.figure(dpi=200)
    ax  = fig.add_subplot()
    ax.plot(x_vals)
```
