```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scienceplots
import sys
sys.path.append("../../../../../src/")
import nanonets_utils
import nanonets

plt.style.use(["science","bright","grid"])

from scipy.optimize import curve_fit
from matplotlib.colors import LogNorm
from scipy.interpolate import interp1d

def sin_func(t, A, f0, phi, B):
    return A*np.sin(2*np.pi*f0*t-phi) + B

def exp_decay(x, a, b, c):
    return a*np.exp(-x/b) + c
```

## Closed Circuit

```python
N_periods   = 40
time_step   = 1e-9
U0          = 0.02
amplitudes  = [U0,0.0]
Np          = 40
N_eq_p      = 20

topology_parameter  = {
    "Nx"                : Np,
    "Ny"                : 1,
    "Nz"                : 1,
    "e_pos"             : [[0,0,0],[Np-1,0,0]],
    "electrode_type"    : ['constant','floating']
}

# folder      = "/mnt/c/Users/jonas/Desktop/phd/data/2_funding_period/current/wo_magic_cable/ac_input_vs_freq/"
folder      = "/home/j/j_mens07/phd/data/2_funding_period/potential/wo_magic_cable/ac_input_vs_freq/"
# freq_vals   = [f0*1e6 for f0 in [1,2,3,4,5,6,7,8,9]]
freq_vals   = [f0*1e6 for f0 in [0.5,0.6,0.7,0.8,0.9,1.0,1.1,1.2,1.3,1.4,1.5,
                                 1.6,1.7,1.8,1.9,2.0,2.1,2.2,2.3,2.4,2.5,2.6,
                                 2.7,2.8,2.9,3.0,3.1,3.2,3.3,3.4,3.5,3.6,3.7,3.8,3.9,4.0]]
time_data   = {}
U0_data     = {}
data        = {}
pot_data    = {}
for f0 in freq_vals:
    frequencies         = [f0,0.0]
    T_sim               = N_periods/f0
    N_volt              = int(T_sim/time_step)
    time_steps, volt    = nanonets_utils.sinusoidal_voltages(N_volt, topology_parameter, amplitudes=amplitudes, frequencies=frequencies, time_step=time_step)
    data[f0]            = pd.read_csv(f"{folder}Nx={Np}_Ny=1_Nz=1_Ne=2_{f0}.csv").iloc[N_eq_p*N_volt//N_periods:,:].reset_index(drop=True)
    pot_data[f0]        = pd.read_csv(f"{folder}mean_state_Nx={Np}_Ny=1_Nz=1_Ne=2_{f0}.csv").iloc[N_eq_p*N_volt//N_periods:,:].reset_index(drop=True)
    time_data[f0]       = time_steps[:len(data[f0])]
    U0_data[f0]         = volt[N_eq_p*N_volt//N_periods:-1,0]
```

```python
N_f         = 50
h_strength  = []
thd         = []
A_f0        = []
phase       = []
freq_fit    = []

for f0 in freq_vals:
    h1  = []
    t1  = []
    A1  = []
    p1  = []
    f1  = []
    for i in range(40):
        t_val   = time_data[f0]
        y_val   = pot_data[f0].iloc[:,i].values.copy()*1000
        h_temp  = nanonets_utils.harmonic_strength(signal=y_val, f0=f0, dt=time_step, N_f=N_f, n_padded=len(y_val)*10, dB=False)
        xf,yf   = nanonets_utils.fft(y_val, time_step, n_padded=len(y_val)*10)
        func    = interp1d(xf, yf)
        A2      = float(func(f0))
        p,_     = curve_fit(sin_func, t_val, y_val, p0=[0.1, f0, 0, 0], bounds=[(0,f0-0.1,-np.inf,-np.inf),(0.2,f0+0.1,np.inf,np.inf)])

        h1.append(h_temp)
        t1.append(np.sqrt(np.sum(np.array(h_temp)**2))*100)
        A1.append(A2)
        f1.append(p[1]/1e6)
        p1.append(p[2])
    A_f0.append(A1)
    h_strength.append(h1)
    thd.append(t1)
    phase.append(p1)
    freq_fit.append(f1)

thd_df  = pd.DataFrame(thd, index=np.round(np.array(freq_vals)*1e-6,2))
A_f0_df = pd.DataFrame(A_f0, index=np.round(np.array(freq_vals)*1e-6,2))
pha_df  = pd.DataFrame(phase, index=np.round(np.array(freq_vals)*1e-6,2))
# h_df    = pd.DataFrame(h_strength, columns=np.arange(2,N_f+2), index=f_vals)
# thd_v   = np.sqrt(np.sum(np.array(h_strength)**2, axis=1))*100
# h_df
```
