```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
e   = 1.60217663*1e-19
kB  = 1.38064852*1e-23

def free_energy(U, C):

    return e*U + (e**2)/(2*C)

def tunneling_rate(dF, R, T=1):

    a   = R*e**2
    b   = -dF / a
    c   = 1 - np.exp(dF/(kB*T))

    return b / c
```

```python
U   = 0.1
C   = 2.24*1e-18
R   = 25*1e6
T   = np.arange(5,20)
dF  = free_energy(U, C)

kBT     = kB*T
dF_kbT  = dF/kBT
gamma   = tunneling_rate(dF, R, T=T)
```

```python
fig = plt.figure(dpi=200)
fig.set_figwidth(fig.get_figwidth()*2)
ax  = fig.add_subplot(1,3,1)
ax.plot(T, dF_kbT)
ax  = fig.add_subplot(1,3,2)
ax.plot(T, np.exp(dF_kbT))
ax.set_yscale('log')
ax  = fig.add_subplot(1,3,3)
ax.plot(T, gamma)
ax.set_yscale('log')
```
