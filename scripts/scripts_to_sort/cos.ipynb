```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N       = 10
con_arr = np.random.randint(low=0,high=2,size=(N,N))
con_arr
```

```python
max_jumps = np.max(np.sum(con_arr, axis=1))

final_list = []
for i in range(con_arr.shape[0]):
    con_list = []
    for j in range(con_arr.shape[1]):
        if con_arr[i,j] == 1:
            con_list.append(j)
    while len(con_list) < max_jumps:
        con_list.append(0)
    final_list.append(con_list)

np.array(final_list)
```

```python
N_voltages      = 200
max_time        = 1e-4
time_steps      = np.linspace(0,max_time,N_voltages)
amplitude       = 0.05
frequency       = 2e5
voltages_values = amplitude*np.cos(frequency*time_steps)
plt.plot(time_steps, voltages_values)
```
