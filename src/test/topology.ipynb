```python
import networkx as nx
import matplotlib.pyplot as plt
import sys
sys.path.append("../")
import topology
```

# Lattice Networks

```python
# Parameter and Inti
N_x, N_y        = 5,3
electrode_pos   = [[0,0],[2,0],[0,2],[2,2]]
lattice         = topology.NanoparticleTopology()

# Build Network and attach Electrodes
lattice.cubic_network(N_x, N_y)
lattice.set_electrodes_based_on_pos(electrode_pos)
print(lattice)
```

```python
G   = lattice.return_graph_object()
pos = lattice.return_np_pos()

fig, ax = plt.subplots(dpi=150)
nx.draw_networkx(G=G, pos=pos, ax=ax)
```

### Option to a NP to Output

```python
# Parameter and Inti
N_x, N_y, N_z   = 3,3,1
electrode_pos   = [[0,0,0],[2,0,0],[0,2,0],[2,2,0]]
lattice2        = topology.topology_class()

# Build Network and attach Electrodes
lattice2.cubic_network(N_x, N_y, N_z)
lattice2.set_electrodes_based_on_pos(electrode_pos, N_x, N_y)
lattice2.add_np_to_output()
print(lattice2)
```

```python
G2      = lattice2.return_graph_object()
pos2    = lattice2.return_np_pos()

fig, ax = plt.subplots(dpi=150)
nx.draw_networkx(G=G2, pos=pos2, ax=ax)
```

# Disorderd Networks

```python
N_p, N_j        = 20,0
electrode_pos   = [[-1,-1],[1,-1],[-1,1],[1,1]]
random_topology = topology.topology_class()
random_topology.random_network(N_p, N_j)
random_topology.add_electrodes_to_random_net(electrode_pos)
random_topology.graph_to_net_topology()
print(random_topology)
```

```python
Gr      = random_topology.return_graph_object()
posr    = random_topology.return_np_pos()

fig, ax = plt.subplots(dpi=150)
nx.draw_networkx(G=Gr, pos=posr, ax=ax)
```
