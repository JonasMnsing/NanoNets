```python
import numpy as np
```

```python
def mutal_capacitance_adjacent_spheres(eps_r : float, np_radius : float, np_distance : float)->float:

    factor  = 4*3.14159265359*8.85418781762039*0.001*eps_r
    cap     = factor*((np_radius**2)/(2*np_radius + np_distance))

    return cap

eps_r=2.6
eps_s=3.9
np_radius=10.0
np_distance=1.0

C_lead  = mutal_capacitance_adjacent_spheres(eps_r, np_radius, np_distance)
C_lead
```

```python
volt_configs = np.loadtxt("voltage_configs.csv")
volt_configs = volt_configs*C_lead
volt_configs
```

```python
ele_charge     = 0.160217662

for i in range(200):
    
    try:
        arr = np.loadtxt(f"{i}")
        a   = volt_configs[i,0]
        b   = volt_configs[i,1]
        c   = volt_configs[i,2]
        d   = volt_configs[i,3]

        arr[:,0]    = (arr[:,0]  - a)
        arr[:,3]    = (arr[:,3]  - b)
        arr[:,12]   = (arr[:,12] - c)
        arr[:,15]   = (arr[:,15] - d)

        arr = arr/ele_charge
        arr = np.round(arr)

        np.savetxt(f"new/{i}_.csv", arr)
    except:
        pass
```
